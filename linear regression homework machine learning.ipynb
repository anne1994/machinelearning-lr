```python
%matplotlib inline
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
from pylab import randn
from numpy import random
random.seed(30)
x= randn(30)
#print("x",x)
y=randn(30)
#print("y",y)
plt.scatter(x,y,c=['r','b'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('linear regression')
plt.show()
```

```python
def calculate(i,o):
    n=len(i)
    xmean=np.mean(i , None)
    ymean=np.mean(o, None)
    ssxm, ssxym, ssyxm, ssym = np.cov(i, o, bias=1).flat
    slope=ssxym/ssxm
    #print("slope is",slope)
    intercept = ymean-slope*xmean
    #print("intercept is " , intercept)
    return (slope,intercept)
```

```python
slope,intercept=calculate(x,y)
print("slope" , slope)
print("intercept", intercept)
```

```python
def prediction(x):
    return slope * x + intercept

line = prediction(x)
#y = np.where(np.array(y))
plt.xlabel('x')
plt.ylabel('y')
plt.title('linear regression')

#plt.scatter(x,y,c=y, s=150, marker='<', edgecolor='none')
plt.scatter(x,y,c=['r','b'])
plt.plot(x,line, c='r')
plt.show()
```
